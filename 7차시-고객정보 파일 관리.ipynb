## 1. 고객정보 파일 분할

### 1.1 함수 정의 및 사용

```python
import pandas as pd
```

```python
고객정보 = pd.read_excel('../데이터/고객정보.xlsx')
```

```python
고객정보
```

#### 엑셀저장 함수 정의

```python
def 엑셀저장(데이터프레임, 경로명):
    엑셀 = pd.ExcelWriter(경로명)
    데이터프레임.to_excel(엑셀, 경로명, index = False)
    엑셀.save()
```

#### 엑셀저장 함수 사용

```python
고객정보_서울시 = 고객정보[고객정보.주소.str.startswith('서울시')]
고객정보_서울시
```

```python
엑셀저장(고객정보_서울시, '고객정보_서울시.xlsx')
```

### 1.2 고객정보 파일 분할

#### 주소의 도시에 따라 파일 분할

```python
도시목록 = ['서울시', '대전시', '부산시', '인천시', '광주시', '대구시', '제주시']
```

```python
도시목록
```

```python
for 시명 in 도시목록:
    print(시명)
```

```python
for 시명 in 도시목록:
    파일명 = '고객정보_{}.xlsx'.format(시명)
    print(파일명)
```

#### 고객정보 파일분할

- 1단계 : 고객정보 검색 
          고객정보_서울시 = 고객정보[고객정보.주소.str.startswith('서울시')]
- 2단계 : 파일명 생성 
          파일명 = '고객정보_{}.xlsx'.format(시명)
- 3단계 : 엑셀저장 함수 호출
          엑셀저장(고객정보_서울시, '고객정보_서울시.xlsx')

```python
for 시명 in 도시목록:
    고객정보_도시별 = 고객정보[고객정보.주소.str.startswith(시명)]    
    파일명 = '고객정보_{}.xlsx'.format(시명) 
    엑셀저장(고객정보_도시별, 파일명)
    print(파일명 + ' 저장완료...')
```

### 2. 고객정보 파일 통합

### 2.1 통합 파일목록 생성

```python
import os
```

```python
검색디렉토리 = '.'
for 경로명, 디렉토리목록, 파일목록 in os.walk(검색디렉토리):
    for 파일명 in 파일목록:
        print(경로명 + '-' + 파일명)
```

#### 통합 파일목록 생성

```python
통합파일목록 = []
for 경로명, 디렉토리목록, 파일목록 in os.walk(검색디렉토리):
    for 파일명 in 파일목록:
        if 파일명.endswith('.xlsx'):
            통합파일목록.append(파일명)
            print(파일명)        
```

```python
통합파일목록
```

### 2.2 고객정보 파일 통합

- 통합파일 목록 생성
- for 구문 사용

- 엑셀파일 읽어서 데이터프레임 저장
- 데이터프레임을 리스트에 저장

- 리스트를 데이터 프레임으로 변환

#### 고객정보 파일 통합

```python
데이터프레임_리스트 = []
for 파일 in 통합파일목록:
    데이터프레임 = pd.read_excel(파일)
    데이터프레임_리스트.append(데이터프레임)
    
고객정보_통합 = pd.concat(데이터프레임_리스트)
```

```python
고객정보_통합
```

#### 인덱스 재생성

```python
고객정보_통합 = 고객정보_통합.reset_index(drop=True)
고객정보_통합
```

### 2.3 고객정보 파일 정렬

- sort_value 함수 이용
- 데이터프레임.sort_value(['컬럼명'], ascending=[True])
- 순정렬 : ascending = [True], 역정렬 : ascending = [False]
- 컬럼명은 리스트 형태로 여러 개를 지정할 수 있음
- 컬럼명이 복수개 일 경우, ascending 값을 동일한 수로 지정 필요

#### 단일 컬럼 순정렬

```python
고객정보_통합.sort_values(['고객번호'], ascending=[True])
```

#### 단일컬럼 역정렬

```python
고객정보_통합.sort_values(['고객번호'], ascending=[False])
```

#### 복수컬럼 정렬

```python
고객정보_통합.sort_values(['성명', '주소'], ascending=[True, True])
```

```python
엑셀저장(고객정보_통합, '고객정보_통합.xlsx')
```
