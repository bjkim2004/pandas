## 1. 고객정보 쉬트 분할 및 통합

### 1.1 고객정보 쉬트 분할

#### 고객정보 읽어오기

```python
import pandas as pd
```

```python
고객정보파일 = '../데이터/고객정보.xlsx'
```

```python
고객정보 = pd.read_excel(고객정보파일)
```

```python
고객정보
```

#### 엑셀저장 3단계 중 2단계를 반복 수행

- 1단계 : 엑셀 = pd.ExcelWriter(파일명)
- 2단계 : 데이터프레임.to_excel(엑셀, '쉬트명', index = False)
- 3단계 : 엑셀.save()

#### 도시별 쉬트 저장

```python
도시목록 = ['서울시', '인천시', '대전시', '광주시', '대구시', '부산시', '제주시']
```

```python
쉬트분할엑셀 = pd.ExcelWriter('고객정보_쉬트분할.xlsx')

for 도시 in 도시목록:
    데이터프레임 = 고객정보[고객정보.주소.str.startswith(도시)]
    데이터프레임.to_excel(쉬트분할엑셀, 도시, index=False)  

쉬트분할엑셀.save()
```

#### 리스트를 이용한 쉬트 통합

```python
데이터프레임_리스트 = []
for 인덱스 in range(0, 7):
    데이터프레임 = pd.read_excel('고객정보_쉬트분할.xlsx', sheetname=인덱스)
    데이터프레임_리스트.append(데이터프레임)   
    
데이터프레임_통합 = pd.concat(데이터프레임_리스트)
```

```python
데이터프레임_통합
```

### 2.2 고객정보 신규 컬럼 추가
